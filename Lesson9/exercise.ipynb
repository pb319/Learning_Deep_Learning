```python
# Instatiating Class Variable
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
scale = MinMaxScaler() 
label = OneHotEncoder()
```

```python
#Labelling Gender Column
df.Gender.replace({"Female":0,"Male":1},inplace= True)
```

```python
dumm1 = pd.get_dummies(df[["Geography"]],drop_first= True,dtype= int)
df.drop(["Geography"],axis= "columns",inplace= True)
pd.concat([df,dumm1], axis= 1)
```

```python
#Standardized Features
col = ['CreditScore','Age', 'Tenure', 'Balance','EstimatedSalary','NumOfProducts']
for i in col:
    df[i] = (df[i]-df[i].min())/(df[i].max()-df[i].min())
```

```python
# Segregating Features and Target
X = df.drop(["Exited"],axis="columns")
y = df["Exited"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_t, y_train, y_t  =train_test_split(X,y,test_size= 0.20)
X_cv, X_test, y_cv, y_test  =train_test_split(X_t,y_t,test_size= 0.60)
```

```python
X_train.shape,X_cv.shape,X_test.shape,y_train.shape,y_cv.shape,y_test.shape
```

```python
X_train.shape
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.optimizers import Adam
```

```python
layer1 = Dense(units=20,input_shape=(9,),activation="relu")
layer2 = Dense(units=15,activation="relu")
layer3 = Dense(units=15,activation="relu")
layer4 = Dense(units=1,activation="sigmoid")

model1 = Sequential([layer1, layer2, layer3, layer4])
model2 = Sequential([layer1, layer2, layer3, layer3, layer3, layer4])
model3 = Sequential([layer1, layer2, layer3, layer3,layer4])


model1.compile(
    optimizer= Adam(),
    loss= BinaryCrossentropy(),
    metrics= ["accuracy"]
)

model2.compile(
    optimizer= Adam(),
    loss= BinaryCrossentropy(),
    metrics= ["accuracy"]
)
model3.compile(
    optimizer= Adam(),
    loss= BinaryCrossentropy(),
    metrics= ["accuracy"]
)
```

```python
#fitting all three models with same epoch
model1.fit(X_train,y_train,epochs= 100)
model2.fit(X_train,y_train,epochs= 100)
model3.fit(X_train,y_train,epochs= 100)
```

```python
#evaluating on Development set
model1.evaluate(X_cv,y_cv), model2.evaluate(X_cv,y_cv), model3.evaluate(X_cv,y_cv)
```

```python
y_pred = []
for i in model3.predict(X_test):
    if i >= 0.50:
        y_pred.append(1)
    else:
        y_pred.append(0)
print(y_pred)
```
