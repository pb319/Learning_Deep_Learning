```python
import math as m
def sigmoid(x):
    return 1/(1+m.exp(-x))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
lst = [sigmoid(i) for i in range(-20,20)]
plt.scatter(range(-20,20), lst , alpha= 0.35)
plt.plot(range(-20,20),lst, color ="black", alpha = 0.50)
```

```python
def tanh(x):
    a = m.exp(x)
    b =m.exp(-x)
    return((a-b)/(a+b))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
lst = [tanh(i) for i in range(-20,20)]
plt.scatter(range(-20,20), lst , alpha= 0.35)
plt.plot(range(-20,20),lst, color ="black", alpha = 0.50)
```

```python
def ReLU(x):
    return max(0,x)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
lst = [ReLU(i) for i in range(-20,20)]
plt.scatter(range(-20,20), lst , alpha= 0.35)
plt.plot(range(-20,20),lst, color ="black", alpha = 0.50)
```

```python
def LReLU(x):
    return max(0.1*x,x)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
lst = [LReLU(i) for i in range(-20,20)]
plt.scatter(range(-20,20), lst , alpha= 0.35)
plt.plot(range(-20,20),lst, color ="black", alpha = 0.50)
```
