```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/home/prasun/GitDemo/Learning_Deep_Learning/Lesson 10/sonar.all-data.xls")
df.head()
```

```python
#Basic Exploration
df.info()
```

```python
# Lebelling Target Variable
df.Label.replace({"M":1,"R":0},inplace= True)
```

```python
#Segregating Data
X = df.drop(["Label"],axis= "columns")
y = df[["Label"]]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25)
```

```python
X_train.shape
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import Sequential
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.optimizers import Adam
```

```python
layer1 = Dense(units=50,input_shape=(60,),activation= "relu" )
layer2 = Dense(units=40,activation= "relu" )
layer3 = Dense(units=40,activation= "relu" )
layer4 = Dense(units=45,activation= "relu" )
layer5 = Dense(units=1,activation= "sigmoid")

model1 = Sequential([layer1, layer2, layer3, layer4, layer5])

model1.compile(
    optimizer = Adam(),
    loss = BinaryCrossentropy(),
    metrics = ["accuracy"]
)
```

```python
model1.fit(X_train,y_train,epochs= 200,batch_size= 8)
```

```python
model1.evaluate(X_test,y_test)
```

```python
model1.summary()
```

```python
layer1 = Dense(units=50,input_shape=(60,),activation= "relu" )
layer2 = Dense(units=40,activation= "relu" )
layer3 = Dense(units=40,activation= "relu" )
layer4 = Dense(units=45,activation= "relu" )
layer5 = Dense(units=1,activation= "sigmoid")
layerD = Dropout(0.5)

model2 = Sequential([layer1, layerD, layer2, layerD, layer3, layerD, layer4, layerD, layer5])

model2.compile(
    optimizer = Adam(),
    loss = BinaryCrossentropy(),
    metrics = ["accuracy"]
)
```

```python
model2.fit(X_train,y_train,epochs= 200, batch_size= 8)
```

```python
model2.summary()
```

```python
model2.evaluate(X_test,y_test)
```
