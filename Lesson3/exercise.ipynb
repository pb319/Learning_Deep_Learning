```python
import numpy as np
```

```python
x_test = np.array([4,5,6,7,8])
y_hat = np.array([4,5,6,7,8])
y_true = np.array([4.81,4.71,7,7.91,8.05])
```

```python
# MSE
def mse(y_true1, y_hat1):
    cost = np.mean((y_true1-y_hat1)**2)
    return cost
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize= (4,3))
plt.xlabel("Independent")
plt.ylabel("Dependent")
plt.scatter(x_test,y_true, color = "Red")
plt.plot(x_test,y_hat)
```

```python
print(mse(y_true,y_hat))
```
