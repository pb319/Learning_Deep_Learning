```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['age','affordibility']],df.bought_insurance,test_size=0.2, random_state=25)
```

```python
X_train_scaled = X_train.copy()
X_train_scaled['age'] = X_train_scaled['age'] / 100

X_test_scaled = X_test.copy()
X_test_scaled['age'] = X_test_scaled['age'] / 100
```

```python
# numpy induced sigmoid function
def signew(X):
    return 1/(1+np.exp(-X))
```

```python
# Logloss Function
def logloss(y_true,y_hat):
    epsilon = 1e-15
    y_predicted = [max(i,epsilon) for i in y_hat]
    y_predicted = [min(i,1-epsilon) for i in y_hat]
    ypred = np.array(y_predicted)
    return (-np.mean((y_true*np.log(ypred))+((1-y_true)*np.log(1-ypred))))
```

```python
def Gradient_Descent(age,affordibility,y_true,epoch):
    # w1,w2,bias are the output expected from the function
    w1= w2= 1
    bias = 0
    l_rate = 0.05
    n = len(age)

    for i in range(epoch):
        wsum = w1*age + w2*affordibility + bias
        ypred = signew(wsum)
        loss = logloss(y_true,ypred)

        # w1= w1 - (l_rate*(np.mean((ypred-y_true)*age)))
        # w2= w2 - (l_rate*(np.mean((ypred-y_true)*affordibility)))
        # w1= w1 - (l_rate*(np.mean((ypred-y_true))))

        w1d = (1/n)*np.dot(age,(ypred-y_true))
        w2d = (1/n)*np.dot(affordibility,(ypred-y_true))
        biasd = (1/n)*np.mean(ypred -y_true)

        w1 = w1 - l_rate *w1d
        w2 = w2 - l_rate *w2d
        bias = bias - l_rate *biasd

        # w1= w1 - (l_rate*((1/n)*np.dot(np.transpose(age),(ypred-y_true))))
        # w2= w2 - (l_rate*((1/n)*np.dot(np.transpose(affordibility),(ypred-y_true))))
        # bias = bias - (l_rate*(np.mean((ypred-y_true))))

        print(f"epoch:{i}, w1:{w1}, w2:{w2} ,b:{bias},loss:{loss}")
    return w1,w2,bias
```

```python
Gradient_Descent(X_train_scaled['age'],X_train_scaled['affordibility'],y_train,1000)
```
